```python
import csv
import matplotlib.pyplot as plt

# Read the data from the CSV file
usage_data = []
users_data = []

with open('data.csv', 'r') as file:
    reader = csv.reader(file)
    next(reader)  # Skip the header row

    for row in reader:
        usage_data.append(float(row[1]))
        users_data.append(int(row[2]))

# Create the diagram
plt.plot(users_data, usage_data)
plt.xlabel('Users')
plt.ylabel('Usage')

plt.title('Usage vs. Users')

plt.show()
```
